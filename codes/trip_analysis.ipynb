# Aggregate data into 'trip' level

Method:  
1. Group orders by `[customer_id, pickup_s2id_15, dropoff_s2id_15]`.  
2. Set time threshold of two consecutive orders. Two orders have a `booking_time` gap longer than the threshold cannot be aggregated to one trip.
3. But it is hard to set a single time-threshold to cover all possible situations. For some cases, the threshold might be too long that orders actually belong to multiple trips are aggregated into one, while for others, the threshold is too short that orders belong to one trip are assigned to multiple trips.
4. Hence, we need to add another condition: for each trip, there should be no more than one 'completed' status. This is to ensure that orders with very short `booking_time` gap but belongs to different trips would not be wrongly classify into one trip.
5. Both the 'booking_time threshold' and 'count of completed status' conditions are necessary. If we only have the 'count of completed status' condition, there will be orders with large `booking_time` difference being classified into one trip, which is unreasonable.

```python
import pandas as pd

# Sort the DataFrame by 'customer_id' and 'booking_time'
df_demand.sort_values(by=['customer_id', 'booking_time'], inplace=True)

# Initialize trip_id and trip_start_time
trip_id = 1
trip_start_time = None

# Lists to store trip IDs and indices
trip_ids = []
trip_indices = []

# Iterate through the DataFrame to assign trip IDs
for index, row in df_demand.iterrows():
    if trip_start_time is None:
        # First row of a trip
        has_status_1 = (row['status'] == 1)  # Check if the first row has status 1
    elif (
        prev_row is not None and
        row['customer_id'] == prev_row['customer_id'] and
        row['pickup_s2id_15'] == prev_row['pickup_s2id_15'] and
        row['dropoff_s2id_15'] == prev_row['dropoff_s2id_15'] and
        (row['booking_time'] - trip_start_time) <= pd.Timedelta(minutes=5) and
        not has_status_1  # Check if there is no status 1 in previous rows
    ):
        # Continue the trip
        pass
    else:
        # Start a new trip
        trip_id += 1
        has_status_1 = (row['status'] == 1)  # Update has_status_1 for the new trip

    trip_start_time = row['booking_time']
    trip_ids.append(trip_id)
    trip_indices.append(index)
    prev_row = row

# Assign the 'trip_id' values to the DataFrame
df_demand['trip_id'] = trip_ids

# Save the trip-level dataset to a new CSV file
df_demand.to_csv('/home/xychen/gojeck_data_proc/results/trip_data_status.csv', index=False)
```

## Overlap check
Overlap: for a certain customer, there is another booking time in between the pickup and dropoff time period of the previous orders

```python
from tqdm import tqdm
import pandas as pd

# Sort the original DataFrame by 'booking_time'
df_trip_status = df_trip_status.sort_values(by='booking_time')

# Group by 'trip_id' without creating a separate 'grouped' object
grouped = df_trip_status.groupby('trip_id')

# Get the number of unique groups
num_groups = len(grouped.groups)

with tqdm(total=num_groups, desc="Groupby & Apply Progress") as pbar:
    def overlap(df):
        n_overlap = 0
        prev_pu = None
        prev_do = None

        for index, row in df.iterrows():
            if prev_pu is not None and prev_do is not None:
                # Check if there's an overlap between 'booking_time' and the previous time interval
                if not pd.isna(row['pickup_time']) and not pd.isna(row['drop_off_time']):
                    if (row['booking_time'] >= prev_pu) and (row['booking_time'] <= prev_do):
                        n_overlap += 1

            prev_pu = row['pickup_time']
            prev_do = row['drop_off_time']
        
        pbar.update(1)
        return n_overlap

    results = grouped.apply(overlap)
```

Summary:  

There is not any overlap (booking_time within the time period of previous pickups to dropoffs) in the aggregated trip data using the method aboved.

However, if the condition of 'no more than one "completed" status' is not used in trip aggregation, would there be any overlap?   
--- No. (The following codes show the validation.)

## Time period

```python
from datetime import time
from tqdm import tqdm

# Define the start and end times for time periods outside the function
midnight_start = time(0, 0)
midnight_end = time(6, 0)
am_peak_start = time(6, 0)
am_peak_end = time(10, 0)
day_nonpeak_start = time(10, 0)
day_nonpeak_end = time(17, 0)
# pm_peak_start = time(17, 0)
# pm_peak_end = time(23, 59)

# Define the custom function for time period categorization
def time_period_define(group):
    booking_time = group['booking_time'].time()
    
    if midnight_start <= booking_time < midnight_end:
        return 'midnight'
    elif am_peak_start <= booking_time < am_peak_end:
        return 'am_peak'
    elif day_nonpeak_start <= booking_time < day_nonpeak_end:
        return 'day_nonpeak'
    else:
        return 'pm_peak'

with tqdm(total=len(df_trip_status), desc="Iterating through rows") as pbar:
    df_trip_status['time_period'] = df_trip_status.apply(time_period_define, axis=1)
    pbar.update(len(df_trip_status))
```

## Weekend

```python
def weekend_check(group):
    booking_time = group['booking_time']
    # Use the weekday() method to get the day of the week (0 = Monday, 6 = Sunday)
    day_of_week = booking_time.weekday()
    # Check if the day of the week is either Saturday (5) or Sunday (6)
    if day_of_week == 5 or day_of_week == 6:
        return 1
    else:
        return 0
    
with tqdm(total=len(df_trip_status), desc="Iterating through rows") as pbar:
    df_trip_status['weekend'] = df_trip_status.apply(weekend_check, axis=1)
    pbar.update(len(df_trip_status))
```

## Waiting time

If the trip is never made, the waiting time is NaN.

```python
# import tqdm as tqdm

grouped = df_trip_status.groupby('trip_id')
num_groups = len(grouped)

# with tqdm(total=num_groups, desc="Groupby & Apply Progress") as pbar:

    # Define a custom function to calculate the time gap
def calculate_wait_time(group):
    if group['n_trip'].iloc[0] == 1:
        first_booking_time = group['booking_time'].iloc[0]
        last_pickup_time = group['pickup_time'].iloc[-1]
        time_gap = last_pickup_time - first_booking_time
        # pbar.update(1)
        return time_gap
    else:
        # pbar.update(1)
        return pd.NaT  # Return NaN if n_trip is not 1

# Group by 'trip_id' and apply the custom function
result = grouped.apply(calculate_wait_time)

# Reset the index to obtain a Series
result = result.reset_index(level=0, drop=True)


```

```python
# append to the original data frame
df_waittime = result.reset_index()
df_waittime['trip_id'] = df_waittime['index'] + 1
df_waittime = df_waittime.drop('index',axis=1)
df_waittime.columns = ['wait_time','trip_id']
df_trip_status = df_trip_status.merge(df_waittime, on='trip_id', how='left')
# df_trip_status.to_csv(r'/home/xychen/gojeck_data_proc/results/trip_data_status.csv',index=False)
```

## Travel time
-  '`dropoff time - pickup time`' for the completed order.
- If the trip is never made, the travel time is NaN.

```python
grouped = df_trip_status.groupby('trip_id')
num_groups = len(grouped)

# with tqdm(total=num_groups, desc="Groupby & Apply Progress") as pbar:


def calculate_travel_time(group):
    if group['n_trip'].iloc[0] == 1:
        last_pickup_time = group['pickup_time'].iloc[-1]
        last_dropoff_time = group['drop_off_time'].iloc[-1]
        time_gap =  last_dropoff_time - last_pickup_time# the last order is the completed order if the trip is made successfully
        # pbar.update(1)
        return time_gap
    else:
        # pbar.update(1)
        return pd.NaT  # Return NaN if n_trip is not 1

# Group by 'trip_id' and apply the custom function
result = grouped.apply(calculate_travel_time)

# Reset the index to obtain a Series
result = result.reset_index(level=0, drop=True)
```

```python
# append to the original data frame
df_travel_time = result.reset_index()
df_travel_time['trip_id'] = df_travel_time['index'] + 1
df_travel_time = df_travel_time.drop('index',axis=1)
df_travel_time.columns = ['travel_time','trip_id']
df_trip_status = df_trip_status.merge(df_travel_time, on='trip_id', how='left')
# df_trip_status.to_csv(r'/home/xychen/gojeck_data_proc/results/trip_data_status.csv',index=False)
```

## Origin subzone and destination subzone

```python
import geopandas as gpd
from shapely.geometry import Point

# Load administrative boundary data from a shapefile or other data source
gdf_subzone = gpd.read_file('/home/xychen/gojeck_data_proc/data/master-plan-2014-subzone-boundary-web-shp/MP14_SUBZONE_WEB_PL.shp')


# Create a GeoDataFrame from DataFrame coordinates
geometry_pickup = [Point(xy) for xy in zip(df_trip_status['booking_pickup_longitude'], df_trip_status['booking_pickup_latitude'])]
gdf_coords_pickup = gpd.GeoDataFrame(geometry_pickup, columns=['geometry'], crs='EPSG:4326')

geometry_destination = [Point(xy) for xy in zip(df_trip_status['booking_destination_longitude'], df_trip_status['booking_destination_latitude'])]
gdf_coords_destination = gpd.GeoDataFrame(geometry_destination, columns=['geometry'], crs='EPSG:4326')

# Align the crs
gdf_coords_pickup = gdf_coords_pickup.to_crs(gdf_subzone.crs)
gdf_coords_destination = gdf_coords_destination.to_crs(gdf_subzone.crs)

# Perform a spatial join
result_pickup = gpd.sjoin(gdf_coords_pickup, gdf_subzone, op='within')
result_destination = gpd.sjoin(gdf_coords_destination, gdf_subzone, op='within')

# Extract information from the result
subzone_pickup = result_pickup['SUBZONE_N']
subzone_destination = result_destination['SUBZONE_N']



```

```python
# append the result to trip-level dataframe
df_trip_status['origin_subzone'] = subzone_pickup
df_trip_status['dest_subzone'] = subzone_destination
```
